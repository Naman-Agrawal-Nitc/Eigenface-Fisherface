```python
data_path_train1 = 'images/Training images'
data_path_train2 = 'C:/yihuai/lectures/StatisticalPatternRecognition/final project/images(1)/images/Training images'
data_path_test1 = 'images/Test'
data_path_test2 = 'C:/yihuai/lectures/StatisticalPatternRecognition/final project/images(1)/images/Test/Close test'
data_path_test3 = 'C:/yihuai/lectures/StatisticalPatternRecognition/final project/images(1)/images/Test/Open test'
```

```python
row_size = 128
col_size = 128

def get_data(data_folder_path):
    #list to hold all subject faces
    faces = []
    #list to hold labels for all subjects
    labels = []

    #get the images names that are inside the given subject directory
    subject_images_names = os.listdir(data_folder_path)

    #------STEP-3--------
    #go through each image name, read image, 
    #detect face and add face to list of faces
    for image_name in subject_images_names:
        

        #build image path
        #sample image path = training-data/s1/1.pgm
        image_path = data_folder_path + "/" + image_name

        #read image
        image = cv2.imread(image_path, cv2.COLOR_BGR2GRAY)
       
        #add face to list of faces
        faces.append(image)
        #add label for this face
        splits = image_name.split('_', 2)
        label = splits[0]+'_'+splits[1]
        labels.append(label)
        
    return np.array(faces), np.array(labels)
```

```python
x_train1, y_train1 = get_data(data_path_train1)
x_train2, y_train2 = get_data(data_path_train2)

x_test1, y_test1 = get_data(data_path_test1)
x_test2, y_test2 = get_data(data_path_test2)
x_test3, y_test3 = get_data(data_path_test3)
```

```python
def group_count(data):
    result = {}
    for i in np.unique(data):
        temp = (data ==i)
        result[i] = sum(temp)
#     for k in result:
#         print("label=", k, '; count', result[k])
    return result
```
